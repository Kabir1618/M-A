```python
#Creating a Dictionary to store Merger and Acquisition Data, with required Index
from email import header


Ind = ['Particulars','Premium','Share price acquirer', 'Acquisition price of target', 'Acquisition value', 'Cash used',
'Debt used', 'Shares used', 'No. Shares issued by acquirer', 'Pre merger Income of acquirer', 
'Shares outstanding of acquirer', 'EPS of acquirer standalone', 'Post merger EPS calculation', 'Revenue', 'Gross profit', 
'Less: Operating expenses', 'Selling general & admin', 'EBITDA', 'Less:D&A', 'Add : Other Income', 'EBIT', 
'Less:Interest on debt', 'Less: Interest foregone', 'PBT', 'Less: Income tax', 
'Net Income', 'Outstanding Diluted shares of combined company', 'EPS of combined company', 'EPS of acquirer standalone',
 'Accretive/Dilutive', 'Accretion/Dilution', 'Accretion/Dilution (%)']
M_A = pd.Series(index = Ind)
```

```python
def M_AND_A(data, Model, percent):
    #Premium [1:2] = Percent from input
    Model[1:2] = percent
    #Share price acquirer [2:3] = Share price of acquirer from Data  [16:17, :1]
    Model[2:3] = round(data.iloc[16:17, :1], 2)
    #Acquisition price of target[3:4] = percent[1:2]+1 * target share price data[16:17]
    Model[3:4] = round(data.iloc[16:17, 1:2] * (1+Model[1:2][0]),2)
    #Acquisition value [4:5] = number of outstanding shares (data[14:15]) * Acquisition price of the target [3:4]
    Model[4:5] = round((data.iloc[14:15, 1:2] * Model[3:4][0]),2)
    #Cash used [5:6] = Acquisition value [4:5] * percent of cash used as source of fund [24:25]
    Model[5:6] = round(data.iloc[24:25, 1:2].values[0][0] * Model[4:5][0],2)
    Model[6:7] = round(data.iloc[25:26, 1:2].values[0][0] * Model[4:5][0],2)
    Model[7:8] = round(data.iloc[26:27, 1:2].values[0][0] * Model[4:5][0],2)
    Model[8:9] = round(Model[7:8][0]/Model[2:3][0],2)
    Model[9:10] = round(data.iloc[12:13, :1].values[0][0],2)
    Model[10:11] = round(data.iloc[14:15, :1].values[0][0],2)
    Model[11:12] = round(Model[9:10][0]/Model[10:11][0],2)
    Model[12:13] = "-----------------"
    Model[13:14] = round(data.iloc[0:1, 1:2].values[0][0] + data.iloc[0:1, :1].values[0][0],2)
    Model[14:15] = round(data.iloc[1:2, 1:2].values[0][0] + data.iloc[1:2, :1].values[0][0],2)
    Model[15:16] = "-----------------"
    Model[16:17] = round(data.iloc[3:4, 1:2].values[0][0] + data.iloc[3:4, :1].values[0][0],2)
    Model[17:18] = round(Model[14:15][0] - Model[16:17][0],2)
    Model[18:19] = round(data.iloc[5:6, 1:2].values[0][0] + data.iloc[5:6, :1].values[0][0],2)
    Model[19:20] = round(data.iloc[6:7, 1:2].values[0][0] + data.iloc[6:7, :1].values[0][0],2)
    Model[20:21] = round((Model[17:18][0] - (Model[18:19][0]) + Model[19:20][0]),2)
    Model[21:22] = round((data.iloc[21:22, :1] * Model[6:7][0]),2)
    Model[22:23] = round((data.iloc[20:21, :1] * Model[5:6][0]),2)
    Model[23:24] = round((Model[20:21][0] - (Model[21:22][0] + Model[22:23][0])),2)
    Model[24:25] = round(data.iloc[15:16, :1] * Model[23:24][0] ,2)
    Model[25:26] = round((Model[23:24][0] - Model[24:25][0]), 2)
    Model[26:27] = round((Model[8:9][0] + Model[10:11][0]), 2)
    Model[27:28] = round((Model[25:26][0] / Model[26:27][0]), 2)
    Model[28:29] = Model[11:12][0]
    if Model[27:28][0] > Model[28:29][0] : Model[29:30] = "ACCRETIVE" 
    else : Model[29:30] = "DILUTIVE"
    Model[30:31] = Model[27:28][0] - Model[28:29][0]
    Model[31:32] = round((Model[30:31][0]/ Model[28:29][0])*100, 2)
    return Model
```

```python
MerAqu = pd.DataFrame(index=Ind)
for i in [float(s) for s in input("insert Premiums").split(",")]:
    output = M_AND_A(data, M_A, i)
    MerAqu[str(i*100) + "%"] = output
MerAqu = MerAqu[2:]
MerAqu
```
